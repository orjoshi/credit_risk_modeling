### (Expected Loss) 
`EL = PD * LGD * EAD`

```python
loan_data_preprocessed_new = pd.read_csv('/content/drive/MyDrive/Credit Risk Modelling/loan_data_preprocessed_new.csv')
```

```python
loan_data_preprocessed_new['EL'] = loan_data_preprocessed_new['PD'] * loan_data_preprocessed_new['LGD'] * loan_data_preprocessed_new['EAD']
# (Expected Loss) EL = PD * LGD * EAD.
```

```python
loan_data_preprocessed_new['EL'].describe() # descriptive statisics
```

```python
loan_data_preprocessed_new[['funded_amnt', 'PD', 'LGD', 'EAD', 'EL']].head()
```

```python
loan_data_preprocessed_new['funded_amnt'].describe()  # descriptive statisics
```

```python
loan_data_preprocessed_new['EL'].sum() # Total Expected Loss for all loans
```

```python
loan_data_preprocessed_new['funded_amnt'].sum() # Total funded amount for all loans
```

```python
loan_data_preprocessed_new['EL'].sum() / loan_data_preprocessed_new['funded_amnt'].sum()
# Total Expected Loss as a proportion of total funded amount for all loans.
```
